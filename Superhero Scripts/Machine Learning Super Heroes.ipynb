```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import os
import sklearn
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
%matplotlib inline
```

```python
path = r'/Users/ethanahennes/Desktop/Unit 6/Data'
```

```python
df = pd.read_csv(os.path.join(path,'Super Hero Numerical Only.csv'),index_col = False)
```

```python
df.columns
```

```python
df.plot(x = 'Power Class (1-6)', y='Ablity to Withstand Damage (1-100)',style='o') # The style option creates a scatterplot; without it, we only have lines.
plt.title('Power Class Vs Duriblity')  
plt.xlabel('Power Class (1-6)')  
plt.ylabel('Ablity to Withstand Damage (1-100)')  
plt.show()
```

```python
X = df['Power Class (1-6)'].values.reshape(-1,1)
y = df['Ablity to Withstand Damage (1-100)'].values.reshape(-1,1)
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
regression = LinearRegression()  
```

```python
regression.fit(X_train, y_train)
```

```python
y_predicted = regression.predict(X_test)
```

```python
df.columns
```

```python


plot_test = plt
plot_test.scatter(X_test, y_test, color='gray', s = 15)
plot_test.plot(X_test, y_predicted, color='red', linewidth =3)
plot_test.title('Power Class Vs Duriblity (Test set)')
plot_test.xlabel('Power Class (1-6)')
plot_test.ylabel('Ablity to Withstand Damage (1-100)')
plot_test.show()
```

```python
rmse = mean_squared_error(y_test, y_predicted) 
r2 = r2_score(y_test, y_predicted) 
```

```python
print('Slope:' ,regression.coef_)
print('Mean squared error: ', rmse)
print('R2 score: ', r2)
```
